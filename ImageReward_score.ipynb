```python
import ImageReward as RM
model = RM.load("ImageReward-v1.0")
```

```python
# prompt: display image at ing_paths

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

img_path = "elephant.png"

img = np.asarray(Image.open(img_path))
plt.imshow(img)
plt.title(img_path.split(".")[0])
plt.show()
```

```python
prompts = ["elephant", "painting", "painting of an elephant","green color"]



for prompt in prompts:
	reward = model.score(prompt, img_path)
	print(f"{prompt}: {reward}")
```
